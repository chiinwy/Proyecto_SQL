## Tabla Actor

```python
actor = pd.read_csv('C:\\Users\\chiin\\Desktop\\Proyecto_SQL\\data\\actor.csv', encoding='latin-1', low_memory = False)

actor.head()
```

```python
actor.shape
```

```python
actor.drop_duplicates()

actor.dropna()

actor.shape

#no hay duplicados ni nulos en la tabla actores
```

```python
#vamos a concatenar los nombres en una columna

actor['full_name'] = actor['first_name'] + ' ' + actor['last_name']

actor.head()
```

```python
actor = actor.drop(['first_name', 'last_name'], axis=1)

actor.head()
```

```python
actor = actor.reindex(columns=['actor_id', 'full_name', 'last_update'])

actor.head()
```

```python
actor = actor.drop(['last_update'], axis=1)

actor.head()
```

```python
actor.to_csv('C:\\Users\\chiin\\Desktop\\Proyecto_SQL\\data\\actor_limpio.csv', index=False)
```
